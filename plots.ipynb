```python
import matplotlib.pyplot as plt
import numpy as np


x1 = [1, 1, 2, 2]
x2 = [1, 2, 2, 1]
y = [0, 1, 0, 1]


colors = ['blue' if val == 0 else 'red' for val in y]
plt.scatter(x1, x2, c=colors, label='Data Points')
plt.xlabel('x1')
plt.ylabel('x2')

plt.title('Question 2 Plot')


plt.savefig('Q2.png')

plt.show()
```
